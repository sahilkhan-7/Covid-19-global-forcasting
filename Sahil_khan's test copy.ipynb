```python
df = pd.read_csv(r"D:\Python Projects\ML\COVID 19 Forcasting\train.csv")
```

```python
df = df.drop(columns = ['County', 'Province_State'], axis=1)
```

```python
df[df['Target'] == 'ConfirmedCases']
```

```python
df[df['Target'] == 'ConfirmedCases']['Country_Region'].unique()
```

```python
confirmed_cases_df = df[df['Target'] == 'ConfirmedCases']
confirmed_cases_by_country = confirmed_cases_df.groupby('Country_Region')['TargetValue'].sum()
```

```python
confirmed_cases_by_country
```

```python
country_with_high_cases = confirmed_cases_by_country.idxmax()
highest_confirmed_cases = confirmed_cases_by_country.max()
print(f"The country with the highest number of cases is {country_with_high_cases} with {highest_confirmed_cases} cases.")
```
